```python
import matplotlib.pyplot as plt

#Tendncia de Estoque Dirio - Grfico de Linha
dias = [1, 2, 3, 4, 5, 6, 7]
estoque = [100, 95, 110, 105, 120, 115, 130]

plt.figure(figsize=(6,4))
plt.plot(dias, estoque, marker='o', color='blue', linestyle='-')
plt.title('Tendncia de Estoque Dirio')
plt.xlabel('Dia da Semana')
plt.ylabel('Quantidade em Estoque')
plt.grid(True)
plt.show()

#Comparao de Produtos - Grfico de Barras
produtos = ['Teclado', 'Mouse', 'Monitor', 'Webcam']
quantidades = [50, 75, 30, 60]

plt.figure(figsize=(6,4))
plt.bar(produtos, quantidades, color=['green', 'orange', 'red', 'blue'])
plt.title('Comparao de Produtos em Estoque')
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.show()

#Proporo de Categorias - Grfico de Pizza
categorias = ['Eletrnicos', 'Vesturio', 'Alimentos']
valores = [15000, 8000, 5000]
cores = ['#66b3ff','#99ff99','#ffcc99']

plt.figure(figsize=(6,6))
plt.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=90, colors=cores)
plt.title('Proporo do Valor Total de Estoque por Categoria')
plt.show()

#Preo vs. Quantidade - Grfico de Disperso
precos = [50, 120, 300, 80, 20]
estoque_itens = [80, 25, 10, 70, 150]

plt.figure(figsize=(6,4))
plt.scatter(precos, estoque_itens, color='purple', marker='o')
plt.title('Relao entre Preo e Quantidade em Estoque')
plt.xlabel('Preo Unitrio (R$)')
plt.ylabel('Quantidade em Estoque')
plt.grid(True)
plt.show()
```
